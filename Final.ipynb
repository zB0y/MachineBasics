## Preprocessing ##

```python
df = pd.read_csv("C:\\Users\\zoheb\\Dev\\ML Project\\K-nearest Neighbor\\consumer-reviews-of-amazon-products\\1429_1.csv")
df
```

```python
df = df[['reviews.text', 'reviews.rating']]
df = df.drop_duplicates()
df = df.dropna()
df
```

### Balancing Data Frame

```python
notFive = df.drop(df[df['reviews.rating'] == 5.0].index)
notFive = notFive.reset_index(drop=True)
notFive
```

```python
NewnotFive = pd.DataFrame()
for i in range(3):
    NewnotFive = pd.concat([NewnotFive,notFive], axis = 0)
df = pd.concat([df,NewnotFive], axis = 0)
#df = pd.concat([df,notFive], axis = 0)
df
```

```python
import re
NON_ALPHANUM = re.compile(r'[\W]')
NON_ASCII = re.compile(r'[^a-z0-1\s]')
def normalize_texts(texts):
    normalized_texts = []
    for text in texts:
        lower = text.lower()
        no_punctuation = NON_ALPHANUM.sub(r' ', lower)
        no_non_ascii = NON_ASCII.sub(r'', no_punctuation)
        normalized_texts.append(no_non_ascii)
    return normalized_texts
        
df['reviews.text'] = normalize_texts(df['reviews.text'])
```

```python
 df
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(binary=True)
cv.fit(df['reviews.text'])
X = cv.transform(df['reviews.text'])
X
```

```python
print(cv.vocabulary_)
```

## K-Nearest Algorithm ##

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(
    X, df['reviews.rating'], train_size = 0.75
)
    
knn = KNeighborsClassifier(n_neighbors=7)
 
knn.fit(X_train, y_train)
print ("Accuracy = %s" % (accuracy_score(y_val, knn.predict(X_val))))
cmForK = confusion_matrix(y_val, knn.predict(X_val))
```

## Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB
mnb = MultinomialNB()
from sklearn.model_selection import train_test_split

mnb.fit(X_train, y_train)
print ("Accuracy = %s" % (accuracy_score(y_val, mnb.predict(X_val))))
cmforNB = confusion_matrix(y_val, mnb.predict(X_val))
```

## Linear SVM

```python
from sklearn.svm import SVC
clf = SVC(kernel='linear')
from sklearn.model_selection import train_test_split

clf.fit(X_train, y_train)
print ("Accuracy = %s" % (accuracy_score(y_val, clf.predict(X_val))))
confusion_matrix(y_val, clf.predict(X_val))
```
